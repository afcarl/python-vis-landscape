## Notebook Setup

## Data

````python
import pandas as pd
data = pd.read_csv('cars.csv')
data.head()
````

## Simple Scatter Plot

````python
from bokeh.io import show
from bokeh.plotting import figure
````

````python
p = figure()
p.circle(data.Horsepower, data.Miles_per_Gallon)
show(p)
````

## Charts API

````python
from bokeh.charts import Scatter
````

````python
p = Scatter(data, x='Horsepower', y='Miles_per_Gallon')
show(p)
````
